## Research Question

<b> How do product attributes such as category, price, color, and season explain variation in sales volume in Zara’s retail dataset? </b>

The Kaggle dataset, ZARA Sales, includes the following columns: Product ID, Product Position, Promotion, Product Category, Seasonal, Sales Volume, brand, url, sku, name, description, price, currency, scraped_at, terms, and section. We will be utilizing most of the prior columns not including url and sku.

We will be utilizing regression coefficients and the ANOVA test to measure and visualize the variance in sales volume. This is a statistical inference task where we're seeking to understand what features have the highest impact on the flucuation of sales, allowing us to understand what types of product to further push in order to optimize sales. 

## Background and Prior Work


Through this project, we're seeking to understand which product features have the biggest impact on the variance of sales for Zara, a leading fast-fashion company. We hope to utilize the Zara Sales dataset on Kaggle in order to perform various tests including regression models and ANOVA to identify which features have the largest weight: https://www.kaggle.com/datasets/xontoloyo/data-penjualan-zara. Ideally, we're able to extrapolate what we find from this data and apply it to the fast fashion, or fashion, industry as a whole.

In Mehdi Benbiba's project, Zara Sales Analysis<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1) she actually uses the same Kaggle dataset that we've chosen and explores univariate and bivariate relationships. Initially I was concerned about the project as it used the same dataset we had decided upon, the project itself was very different only showing correlational data between, at most, bivariate data. Although the data is informative and she does showcase correlational data, I don't think there are any huge takeaways or really a major focus of the data. 

In Mohamed Ibrahim's project, Sales Data Analysis<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2) he analyzed sales data to uncover peak sales months, best-selling products, top revenue-generating cities, and more. The results of his project showed that Monday was the top selling day (annually), December was the top selling month, the most number of orders occurred at 7PM, and more. Although his research shows a lot of great content, there are a number of limitations with what is implicated. For example, the product mix he used was limited to electronic devices including batteries, headphones, and charging cables. Furthermore, when he analyzed the best-selling day of the week, he didn't aggregate the data over the period of the year and instead visualized it by day, month which is a duplicate of the visualization showcasing what the best-selling month was. Finally, he didn't particularly look into whether the outcome of the data was due to causation or simply correlation (i.e. if revenue was so high on Mondays because a ton of Mondays are holidays, etc). 

In this project, done by Nurkholiq Agani Hafid, he performed an analysis on DiCo (Dicoding Collection)<a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3)showcasing what the best-selling product was, what the worst-selling product was, what the demographic of the consumer base was, and more. Although Nurkholiq showcases great insights by identifying the best and worst-case selling product, I think one way he could've taken this a step further would be identifying the reason for it being the best-selling or worst-selling product or discovering what made sales drop in the ensuing months of March. However, it is a great reference point for us to look at for more preliminary analyses.
